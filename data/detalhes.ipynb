```python
def sb_eventos_partida(partida_id):
    eventos = sb.events(match_id=partida_id)
    if 'location' in eventos.columns:
        eventos_validos = eventos[eventos['location'].apply(lambda x: isinstance(x, list) and len(x) == 2)]
        if not eventos_validos.empty:
            eventos_validos.loc[:, ['localizacao_x', 'localizacao_y']] = pd.DataFrame(eventos_validos['location'].tolist(), index=eventos_validos.index)
            eventos = eventos_validos
    return eventos

def identificar_eventos_por_time(eventos_df):
    times = eventos_df['team'].unique()
    eventos_times = {}

    for time in times:
        eventos_time = eventos_df[eventos_df['team'] == time]
        gols = eventos_time[(eventos_time['type'] == 'Shot') & (eventos_time['shot_outcome'] == 'Goal')]
        jogadores_gols = gols['player'].unique()
        assistencias = eventos_time[(eventos_time['type'] == 'Pass') & (eventos_time['pass_outcome'] == 'Assist')]
        jogadores_assistencias = assistencias['player'].unique()
        cartoes_amarelos = eventos_time[eventos_time['type'] == 'Yellow Card']
        cartoes_vermelhos = eventos_time[eventos_time['type'] == 'Red Card']
        desarmes = eventos_time[eventos_time['type'] == 'Tackle']

        eventos_times[time] = {
            "gols": len(gols),
            "jogadores_gols": jogadores_gols,
            "assistencias": len(assistencias),
            "jogadores_assistencias": jogadores_assistencias,
            "cartoes_amarelos": len(cartoes_amarelos),
            "cartoes_vermelhos": len(cartoes_vermelhos),
            "desarmes": len(desarmes)
        }
    return eventos_times

#partida_id = 7585
partida_id = input("Qual  o id da partida? ")

eventos = sb_eventos_partida(partida_id)

eventos_resumo = identificar_eventos_por_time(eventos)

print("Resumo dos eventos principais por time:")
for time, eventos_time in eventos_resumo.items():
    print(f"\nTime: {time}")
    print(f"Gols: {eventos_time['gols']} ({', '.join(eventos_time['jogadores_gols'])})")
    print(f"Assistncias: {eventos_time['assistencias']} ({', '.join(eventos_time['jogadores_assistencias'])})")
    print(f"Cartes Amarelos: {eventos_time['cartoes_amarelos']}")
    print(f"Cartes Vermelhos: {eventos_time['cartoes_vermelhos']}")
    print(f"Desarmes: {eventos_time['desarmes']}")
```

```python
prompt = f"""
{eventos_resumo}
Com base nesses dados, faz uma explicao simples e resumida sobre a partida.
"""

response = model.generate_content(prompt)
print(response.text)
```

```python
def perfil_jogador_por_id(eventos_df, jogador_id):
    eventos_jogador = eventos_df[eventos_df['player_id'] == jogador_id]
    
    passes = eventos_jogador[eventos_jogador['type'] == 'Pass']
    finalizacoes = eventos_jogador[eventos_jogador['type'] == 'Shot']
    desarmes = eventos_jogador[eventos_jogador['type'] == 'Tackle']
    minutos_jogados = eventos_jogador['minute'].max() if not eventos_jogador.empty else 0

    perfil = {
        "jogador_id": jogador_id,
        "passes": len(passes),
        "finalizacoes": len(finalizacoes),
        "desarmes": len(desarmes),
        "minutos_jogados": minutos_jogados
    }
    
    return perfil

partida_id = 7585
eventos = sb_eventos_partida(partida_id)
jogador_id = 3094

perfil = perfil_jogador_por_id(eventos, jogador_id)

perfil_df = pd.DataFrame([perfil])
print("\nPerfil do Jogador (Tabela):")
perfil_df
```
